### Loading a pre-trained YOLO-World moodel and running a prediction on a image

```python
%%time
# Initialize YOLO-World Model
model = YOLOWorld("yolov8l-world.pt")
```

```python
model
```

```python
%%time
# Execute inference with the YOLOv8s-world model on the specified image
results = model.predict("./images/a.jpg", device="cpu", save=True)
```

```python
# Show results
results[0].show()
```

### Set prompts for specific tasks

The YOLO-World framework allows for the dynamic specification of classes through custom prompts, empowering users to tailor the model to their specific needs without retraining.

```python
%%time
# Define custom classes
model.set_classes(["hard hat", "gloves", "vest", "ladder"])
```

```python
# Execute prediction for specified categories on an image
results = model.predict("./images/a.jpg", device="cpu", save=True)
```

```python
# Show results
results[0].show()
```

```python
Image(filename=f"runs/detect/predict/a.jpg", width=600)
```

We can also save a model after setting custom classes. By doing this we can create a version of the YOLO-World model that is specialized for our specific use case.

```python
%%time
# Initialize a YOLO-World model
model = YOLO('yolov8l-world.pt')
```

```python
%%time
model.set_classes(["hard hat", "gloves", "vest", "ladder"])
```

```python
%%time
# Save the model with the defined offline vocabulary
model.save("custom_yolov8l.pt")
```

### After saving, the custom_yolov8l.pt model behaves like any other pre-trained YOLOv8 model but with a key difference: it is now optimized to detect only the classes you have defined.

# Load custom model and perform predictions

```python
%%time
# Load our custom model
model = YOLO('custom_yolov8l.pt')

# Run inference to detect our custom classes
results = model.predict("./images/", device="cpu", save=True)

# Show results
results[0].show()
```
